```python
# Install necessary packages (optional in Colab as most are pre-installed)
# !pip install numpy pandas matplotlib

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. NumPy: Create and manipulate arrays
print("===== NumPy Array Operations =====")
array1 = np.array([10, 20, 30, 40, 50])
array2 = np.arange(1, 6)

print("Array1:", array1)
print("Array2:", array2)
print("Element-wise Sum:", array1 + array2)
print("Mean of Array1:", np.mean(array1))
print("Reshaped Array2 (1x5 -> 5x1):\n", array2.reshape(-1, 1))

# 2. Pandas: Create and use DataFrame
print("\n===== Pandas DataFrame =====")
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'Marks': [85, 63, 77, 90]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Describe basic stats
print("\nSummary Statistics:\n", df.describe())

# Filter rows
print("\nStudents with Marks > 75:\n", df[df['Marks'] > 75])

# 3. Matplotlib: Basic plots
print("\n===== Plotting with Matplotlib =====")
plt.figure(figsize=(8, 5))

# Bar chart for Marks
plt.subplot(1, 2, 1)
plt.bar(df['Name'], df['Marks'], color='skyblue')
plt.title('Student Marks')
plt.xlabel('Name')
plt.ylabel('Marks')

# Line plot for Age
plt.subplot(1, 2, 2)
plt.plot(df['Name'], df['Age'], marker='o', linestyle='-', color='green')
plt.title('Student Ages')
plt.xlabel('Name')
plt.ylabel('Age')

plt.tight_layout()
plt.show()
```
